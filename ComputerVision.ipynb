### Homologar imagenes

En esta sección se realiza el proceso de llevar todas las imágenes a un mismo tamaño.

```python
 carpetas = [
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/battery',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/biological',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/cardboard',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/clothes',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/glass',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/metal',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/paper',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/plastic',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/shoes',
    '/content/drive/MyDrive/1_DCD/M4_1Practica/archive/trash'

]

imagenes=[]
clases=[]
```

Homologamos las imagenes a formato RGB dado que es el que más frecuencia tiene, lo mismo para la extension jpg y la dimension 28 x 28.



La razón del redimensionamiento de las impagenes yace en el hecho de la implementación de *CNN*, ya que podemos impactar al proceso de aprendizaje y por tanto mermar el desempeño.
Se optó por escalar las imágenes a 28 x 28 ya que es la medida de la imagen más pequeña y escalar imágenes hacia un tamaño menor reduce la cantidad de información que se pierde, contrario a lo que sucede cuando hay un incremento en el tamaño. Además de que computacionalmente hablando procesar imágenes más pequeñas es más rápido y por último pero no menos importante,los kernel en las CNN aprenden patrones espaciales que son invariantes al tamaño, así que si se homologa el tamaño, se garantiza que los patrones aprendidos sean aplicables en toda la imagen, independientemente de su tamaño original.


Lo anterior es sin perder de vista que si reescalamos una imagen demasiado grande a un tamaño menor se pueden perder detalles importantes.

```python
for ruta_carp in carpetas:
    print(f"Procesando imágenes en: {ruta_carp}")
    for subdir, dirs, files in os.walk(ruta_carp):
        print(f"Subdirectorio: {subdir}")
        print(f"Número de imágenes: {len(files)}")
        print()
        clase = os.path.basename(subdir)  # Nombre de la carpeta actual como clase
        for img_filename in files:
            img_path = os.path.join(subdir, img_filename)
            img = cv2.imread(img_path)

            if img is not None:
                # Convertir la imagen a formato RGB
                img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # Convierte a JPG y guarda la imagen
                cv2.imwrite(img_path + ".jpg", img)

                # Lee la imagen JPG y redimensiona
                img_jpg = cv2.imread(img_path + ".jpg")
                img_resize = cv2.resize(img_jpg, (28, 28))
                imagenes.append(img_resize)
                clases.append(clase)
            else:
                print(f"La imagen {img_filename} no se pudo leer correctamente.")
```

```python
imagenes= np.array(imagenes)
clases= np.array(clases)
print(imagenes.shape)
print(clases.shape)
```

# 2. Arquitectura

Al menos 8 capas ocultas y 1 convolucional pooling y una capa fully

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
```

```python

model = Sequential()

# Capa Convolucional
model.add(Conv2D(filters=8, kernel_size=4, strides=1, padding='same', input_shape=(28, 28, 3), activation='relu'))

# Capa de Pooling
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Capa Convolucional
model.add(Conv2D(filters=64, kernel_size=4, strides=1, padding='same', activation='relu'))

# Capa de Pooling
model.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Capa Convolucional
model.add(Conv2D(filters=128, kernel_size=4, strides=1, padding='same', activation='relu'))

# Capa de Pooling
model.add(MaxPool2D(pool_size=(2, 2), strides=2))


# Capa de Aplanado
model.add(Flatten())

# CAPA DE DROPOUT
model.add(Dropout(0.25))

# Capa Densa
model.add(Dense(512, activation='elu'))

# Capa Densa
model.add(Dense(256, activation='relu'))

# Capa Densa
model.add(Dense(128, activation='relu'))

# Capa de Salida
model.add(Dense(10, activation='softmax'))
optimizer1 = Adam(learning_rate=0.0001)

# Compilar el modelo
model.compile(loss='categorical_crossentropy', optimizer=optimizer1, metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=3)
```

# 3. Entrenamiento

Para el desarrollo de la arquitectura se realizó la experimentacion variando las capas ocultas, funciones de activación, posición de las capas de dropout para evitar el sobreajuste, el tamaño de los filtros y las épocas, los resultados variaron entre 80% y 90% de accuracy. Abajo se muestra el modelo que tenía un mejor performance.

Se detecta un ligero sobreajuste respecto al conjunto de entrenamiento y validación. Esto debido a que en la función de pérdida tenemos una diferencia de 45 puntos porcentuales entre el conjunto de validacion y el conjunto de entrenamiento y para el accuracy una diferencia 12 puntos porcentuales.

Se definieron 50 épocas, pero gracias al early stop se llego al mejor resultado en siete epocas. Además, es notable que el tiempo de ejecución para entrenar este modelo con la arquitectura actual no fue significativo, lo cual no consumió muchos recursos computacionales.

# 5. Data augmentation

Tras aplicar data augmentation se detecta una degradación en el modelo de la métrica accuracy y la función de pérdida. Asimismo, se observa que el accuracy del conjunto de validación se aleja del obtenido del conjunto de entrenamiento conforme aumenta el número de épocas; este entrenamiento consume mas recursos computacionales. Por otro lado, en la última época se obtiene el mejor accuracy para el train.

Finalmente la brecha entre las métricas del conjunto train y validation es menor en comparación con el modelo base, pero en general dichas métricas no mejoraron.

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from keras.preprocessing.image import ImageDataGenerator
```

```python
data_augmentation = ImageDataGenerator(
    rotation_range=10,  # Rotación aleatoria de la imagen en el rango de 0 a 10 grados
    width_shift_range=0.1,  # Desplazamiento horizontal aleatorio de la imagen
    height_shift_range=0.1,  # Desplazamiento vertical aleatorio de la imagen
    zoom_range=0.3,  # Zoom aleatorio en la imagen
    horizontal_flip=True,  # Volteo horizontal aleatorio de la imagen
    fill_mode='nearest'  # Modo de llenado para píxeles nuevos creados
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import pandas as pd
```

```python
model2 = Sequential()

# Capa Convolucional
model2.add(Conv2D(filters=8, kernel_size=4, strides=1, padding='same', input_shape=(28, 28, 3), activation='relu'))

# Capa de Pooling
model2.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Capa Convolucional
model2.add(Conv2D(filters=64, kernel_size=4, strides=1, padding='same', activation='relu'))

# Capa de Pooling
model2.add(MaxPool2D(pool_size=(2, 2), strides=2))

# Capa Convolucional
model2.add(Conv2D(filters=128, kernel_size=4, strides=1, padding='same', activation='relu'))

# Capa de Pooling
model2.add(MaxPool2D(pool_size=(2, 2), strides=2))


# Capa de Aplanado
model2.add(Flatten())

# CAPA DE DROPOUT
model2.add(Dropout(0.25))

# Capa Densa
model2.add(Dense(512, activation='elu'))

# Capa Densa
model2.add(Dense(256, activation='relu'))

# Capa Densa
model2.add(Dense(128, activation='relu'))

# Capa Densa
#model.add(Dense(64, activation='relu'))

# Capa Densa
#model.add(Dense(32, activation='relu'))

# Capa Densa
#model.add(Dense(16, activation='relu'))

# Capa de Salida
model2.add(Dense(10, activation='softmax'))
optimizer1 = Adam(learning_rate=0.0001)

# Compilar el modelo
model2.compile(loss='categorical_crossentropy', optimizer=optimizer1, metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=3)
```

```python
train_generator = data_augmentation.flow(X_train, y_cat_train, batch_size=32)
valid_generator = ImageDataGenerator().flow(X_valid, y_cat_valid, batch_size=32)

# Definir EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=3)

# Entrenar el modelo con aumento de datos
#%%time
model2.fit(train_generator, epochs=50, validation_data=valid_generator, callbacks=[early_stop])
```
